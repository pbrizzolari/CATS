```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import decomposition
from sklearn.ensemble import RandomForestRegressor
from matplotlib import pyplot
```

```python
data = pd.read_csv("../raw_data/train_call.txt", index_col=0, delimiter="\t").transpose()

topInfo = data.iloc[:3]
data = data.iloc[4:].dropna(axis=1)
data
classes = pd.read_csv("../raw_data/train_clinical.txt", delimiter="\t", index_col=0)
classes
data = classes.join(data).set_index("Subgroup").dropna()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit
import sklearn.feature_selection as feature_selection

lr = linear_model.LogisticRegression(penalty="l2", max_iter=10000)
sss = StratifiedShuffleSplit(n_splits=3,test_size=1/3)
ssi = StratifiedShuffleSplit(n_splits=10,test_size=1/10)
for i in range(10):
    #this is the outer-outer-loop
    
    for train_index, test_index in sss.split(data,data.index):
        #outerloop
        train_data = data.iloc[train_index]
        test_data = data.iloc[test_index]
        for inner_train_index, inner_test_index in ssi.split(train_data, train_data.index):
            #print(inner_test_index.tolist(), test_index.tolist())
            inner_train_data = train_data.iloc[inner_train_index]
            inner_test_data = train_data.iloc[inner_test_index]
            best_features = feature_selection.SelectKBest(feature_selection.chi2, k=150).fit(inner_train_data+1,inner_train_data.index)
            inner_train_data = inner_train_data.transpose().loc[best_features.get_support()].transpose()
            inner_test_data = inner_test_data.transpose().loc[best_features.get_support()].transpose()
            lr.fit(inner_train_data, inner_train_data.index)
            cm = confusion_matrix(inner_test_data.index, lr.predict(inner_test_data), labels=lr.classes_)
            disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=lr.classes_)
            disp.plot()
        
```
